```

#
#Aqui somente oque foi utilizado no site do streamlit para a entrega 2
#de forma resumida para compreeder oque foi utilizado.






import pandas as pd

# ===========================
# LEITURA DOS DADOS
# ===========================
dfCxO = pd.read_csv("CAMPAIGNxORDER.CSV")
dfLJ = pd.read_csv("STORE.csv")
dfSO = pd.read_csv("STOREORDER.csv")


# ===========================
# CRIAÇÃO DATAFRAME LOJAS
# ===========================
dfLJ = pd.DataFrame({
    'Nome_Loja': dfLJ['name'].unique(),
    'ID_Loja': dfLJ['id'].unique()
})


# ===========================
# MERGE LOJAS + CAMPANHAS
# ===========================
dfCxOLJ = pd.merge(
    dfLJ,
    dfCxO,
    left_on='ID_Loja',
    right_on='storeid',
    how='left'
)

dfCxOLJ.drop(columns=['storeid'], inplace=True)


# ===========================
# MÉTRICAS BASE
# ===========================

# vendas campanhas
total_vendas_campanha = dfCxOLJ['order_id'].nunique()

# valor campanhas
total_valor_campanhas = dfCxOLJ.drop_duplicates(subset='order_id')['totalamount'].sum()

# mensagens
total_mensagens = dfCxOLJ['message_id'].count()

# vendas store order
total_vendas = dfSO['id'].nunique()

# valor store order
total_valor_storeorder = dfSO['totalamount'].sum()


# ===========================
# 1. TAXA DE CONVERSÃO (VENDAS)
# ===========================
taxa_conversao_vendas = (total_vendas_campanha / total_vendas) * 100


# ===========================
# 2. TAXA DE CONVERSÃO (VALOR)
# ===========================
taxa_conversao_valor = (total_valor_campanhas / total_valor_storeorder) * 100


# ===========================
# 3. MENSAGENS POR CAMPANHA
# ===========================
mensagens_por_campanha = (
    dfCxOLJ
    .groupby('campaignid')['message_id']
    .count()
    .reset_index()
)


# ===========================
# 4. FATURAMENTO POR LOJA
# ===========================
faturamento_por_loja = (
    dfCxOLJ
    .groupby('Nome_Loja')['totalamount']
    .sum()
    .reset_index()
)


# ===========================
# RESULTADOS
# ===========================
print("===== INDICADORES =====\n")

print(f"Taxa de conversão (vendas): {taxa_conversao_vendas:.2f}%")
print(f"Taxa de conversão (valor): {taxa_conversao_valor:.2f}%\n")
```

```
#
# ===========================
# RESULTADOS
# ===========================

print("Mensagens por campanha:")
print(mensagens_por_campanha)
```

```

# ===========================
# RESULTADOS
# ===========================

print("\nFaturamento por loja:")
print(faturamento_por_loja)
```
